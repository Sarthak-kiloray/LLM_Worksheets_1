Tokenizing with Code, a substitute to  Open AI Tokenizer website "https://platform.openai.com/tokenizer"

import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Hi how are you doing, it's nice to meet you")

tokens

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```
